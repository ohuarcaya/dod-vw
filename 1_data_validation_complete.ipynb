# Validation Web Productivity

```python
import pandas as pd
import numpy as np
import os

def productive_path (tacto, date, hour):
    return "input_data/web_productivity/"+date+"/t0"+tacto+"_c02_"+date+"_"+hour+"_web_productive.csv"


hour_list = ('0'+pd.Series(range(6, 15)).astype(str)).str[-2:] # 06 - 14
tacto_list = ('0'+pd.Series(range(8, 17)).astype(str)).str[-2:] # 08 - 16

evaluation_list=[]

for tacto_ in tacto_list:
    for hour_  in hour_list:
        path = productive_path (tacto=tacto_, date="20200122", hour=hour_)
        if os.path.exists(path):
            df_ = pd.read_csv(path)
            max_sec = df_.time_stamp[df_.index[-1]]
            evaluation_list.append([tacto_, hour_, df_.VA.min(), df_.VA.max(), 
                                    df_.VA.mean(), df_.NVA.mean(), df_.OUT.mean(),
                                    df_.time_stamp[0], max_sec, str(np.floor(max_sec).astype(int)//60)+":"+("0"+str(np.floor(max_sec).astype(int)%60))[-2:]
                                   ])
    
pd.DataFrame(evaluation_list, columns = ["tacto", "hour", "VA_min", "VA_max", "VA_mean", "NVA_mean", "OUT_mean", "interval_tmstmp", "max_tmstmp", "estimated_time"])
```
